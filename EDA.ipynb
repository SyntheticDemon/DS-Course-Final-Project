## Preprocessing

*Most data is either Ordinal to Numinal , We will perform data processing stages to convert these data types to ordinal data*

We will convert the data into suitable float or numerical values to allow model learning and better visualizations 

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import pandas as pd

def convert_nominal_to_ordered_numerical(dataframe, column_name, order):
    order_mapping = {category: idx for idx, category in enumerate(order)}
    dataframe[column_name] = dataframe[column_name].map(order_mapping)
    return dataframe

def convert_nominal_to_numerical(dataframe, column_name):
    label_encoder = LabelEncoder()
    dataframe[column_name] = label_encoder.fit_transform(dataframe[column_name])
    return dataframe, label_encoder
```

#### Convert CALC Levels
